```python
!pip install tinyfish

import os
from tinyfish import TinyFish

# Set your API key
os.environ["TINYFISH_API_KEY"] = "sk-tinyfish-l8woh6lBo8GY70-UPfeQigN56WeiveqH"  # replace with your real key

client = TinyFish()

def analyze_competitor(url):
    print(f"\n===== ANALYZING {url} =====\n")
    final_output = ""

    with client.agent.stream(
        url=url,
        goal="""
        Extract in clean bullet points:
        - Pricing (tiers, amounts)
        - Key product features
        - Target audience
        - Positioning statements
        - Unique selling points
        - Pros and cons (if visible)
        Return ONLY the final structured report.
        """
    ) as stream:

        for event in stream:
            if event.type == "final_result":
                final_output = event.output

    return final_output


# ---- COMPETITOR LIST ----

competitors = [
    "https://notion.so",
    "https://clickup.com"
]

reports = {}

print("\n🔍 Competitor Analysis Report\n")

for c in competitors:
    output = analyze_competitor(c)
    reports[c] = output
    print(output)
```
